### Distribuição Amostral e intervalo de confiança

```python
salarios = dados ['SALARIO']
```

```python
salarios
```

### Média amostral e o desvio amostral

```python
media_amostral = np.mean(salarios)
media_amostral
```

```python
desvio_amostral = np.std(salarios)
desvio_amostral
```

## Defina o nível de confiança:

```python
nivel_confianca = 0.95
```

### Acompanhe o tamanho da amostra:



```python
tamanho_amostra = len(salarios)
tamanho_amostra
```

### Calcule o erro padrão:

```python
from scipy import stats
```

```python
erro_padrao = stats.sem(salarios)
erro_padrao
```

### Calcule o intervalo de confiança:

```python
intervalo_confianca = stats.t.interval(nivel_confianca, tamanho_amostra - 1, loc=media_amostral, scale=erro_padrao)
intervalo_confianca
```

### Featuring engineering

```python
def preencher_nivel(gestor, nivel):
    if gestor == 1 or gestor == 1.0:
        return "Pessoa Gestora"
    else:
        return nivel
```

## Criando a coluna geração:

```python
def determinar_geracao(idade):
  if 39 < idade <= 58:
    return "Geração X"
  elif 29 < idade <= 39:
    return "Geração Millenial"
  elif 13 < idade <= 29:
   return "Geração Z"
  else:
   return "Outra Geração"
```

```python
dados['GERACAO'] = dados ['IDADE'].apply(determinar_geracao)
```

## Criando uma nova coluna de Etnia - brancas, não branca e outras ❤

## FECheck✅

```python
dados['COR/RACA/ETNIA'].value_counts()
```

```python
def determinar_etnia(etnia):
    if etnia == 'Branca':
        return 'Branca'
    elif etnia in ['Preta', 'Parda', 'Indígena']:
        return 'Não Brancas'
    else:
        return 'Outras'
```

```python
dados["ETNIA_DEFINIDA"] = dados['COR/RACA/ETNIA'].apply(determinar_etnia)
```
